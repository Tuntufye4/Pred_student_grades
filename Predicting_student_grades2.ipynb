```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.utils import shuffle
from sklearn.metrics import confusion_matrix
import joblib

```

```python
data = pd.read_csv("student_data.csv")
data.head()
```

```python
data.corr()
```

```python
data = data[["G1", "G2", "G3", "studytime", "failures", "absences"]]
predict = "G3"
x = np.array(data.drop([predict], 1))
y = np.array(data[predict])
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
linear_regression = LinearRegression()
lr = linear_regression.fit(xtrain, ytrain)
```

```python
# save the model to disk
filename = 'finalized_model.sav'
joblib.dump(lr, filename)
```

```python
loaded_model = joblib.load(filename)
accuracy = loaded_model.score(xtest, ytest)

print(accuracy)
```

```python
print('Coefficient: \n', loaded_model.coef_)
print('Intercept: \n', loaded_model.intercept_)
```

```python
predictions = loaded_model.predict(xtest)

for x in range(len(predictions)):
    print(predictions[x], xtest[x], ytest[x])
```
